```python
import pylab as plt
import statsmodels.api as sm
```

```python
lowess = sm.nonparametric.lowess
x = df.Year
y = df.JD
z = lowess(y, x, frac=0.1)
```

```python
plt.plot(df.Year,df.JD, 'k', label='Media Anual', marker='.')
plt.plot(z[:, 0], z[:, 1], 'r-', lw=float(2), label='Media suavizada')
plt.title("Estimados de media global anual combinando Tierra y Oceanos")
plt.xlabel("Ao")
plt.ylabel("Anomalas de temperatura [C]")
plt.legend(loc='upper left')
plt.show()
```
